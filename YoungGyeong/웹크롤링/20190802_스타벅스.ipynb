# 쌤이랑 한거

```python
"""웹브라우저 실행"""
driver = wb.Chrome()

url = "https://www.istarbucks.co.kr/store/store_map.do"
driver.get(url)
```

```python
btn_loca=driver.find_element_by_class_name('loca_search')
btn_loca.click()
```

```python
btn_gwangju = driver.find_element_by_css_selector('ul.sido_arae_box > li + li + li')
btn_gwangju.click()
```

```python
btn_all = driver.find_element_by_css_selector('ul.gugun_arae_box > li')
btn_all.click()
```

```python
soup = bs(driver.page_source, 'html.parser')

"""매장명"""
names = soup.select('ul.quickSearchResultBoxSidoGugun > li > strong')

"""주소"""
addrs = soup.select('ul.quickSearchResultBoxSidoGugun > li > strong + p')
```

```python
name_list = []
addr_list = []
tel_lits = []
for n,a_t in zip(names, addrs):
    name_list.append(n.text)
    addr_list.append(a_t.text[:-12])
    tel_lits.append(a_t.text[-12:])
```

```python
dic = {'매장명':name_list,
      '주소':addr_list,
      '전화번호':tel_lits}
df = pd.DataFrame(dic)
df = df.set_index('매장명')
df
```

# 음료메뉴 영양정보가져오기

```python
"""웹브라우저 실행"""
driver = wb.Chrome()

url = "https://www.istarbucks.co.kr/menu/index.do"
driver.get(url)
```

```python
btn_detail=driver.find_element_by_class_name('menu_drink_btn01')
btn_detail.click()
```

```python
btn_info=driver.find_element_by_class_name('a2')
btn_info.click()
```

```python
soup = bs(driver.page_source, 'html.parser')

"""컬럼ㅇㅇ...."""
col = soup.select('thead > tr > th', limit=7)
col = col[1:7]

"""메뉴정보"""
menu = soup.select('.m_coffee_info > p')
info = soup.select('.m_coffee_info > ul > li > dl > dd')

cnt = 0
de_info = []
for i in range(len(menu)):
    sd_info = []
    for j in range(6):
        sd_info.append(info[cnt].text)
        cnt = cnt+1
    de_info.append(sd_info)
```

```python
col2 = []
for data in col:
    col2.append(data.text)
col2
```

```python
in_menu = []
for menu_data in menu:
    in_menu.append(menu_data.text)
in_menu
```

```python
df = pd.DataFrame(data=de_info, columns=col2, index = in_menu)
df.index.name = '메뉴'
df
```
