# 색상 추가하는 방법

```python
from palettable.colorbrewer.qualitative import Set2_7 as pal
import random
```

```python
pal.colors
```

```python
"""def color_func(**kwargs): 글씨 색깔
    return tuple(pal.colors[random.randint(0,len(pal.colors)-1)])"""

def color_func(**kwargs):
    list1 = [[3,53,157,1],[254,35,112,1],[241,86,49,1],[255,235,40,1]]
    list2 = [[0,65,130,1],[17,141,240,1],[251,255,163,1],[255,75,104,1]]
    list3 = [[0,114,187,1],[255,76,59,1],[255,208,52,1],[51,51,51,1]]
    return tuple(list3[random.randint(0,len(list3)-1)])
```

```python
wc.recolor(color_func=color_func)
```

```python
plt.figure(figsize=(20,20))
plt.imshow(wc)
plt.axis('off')
```

```python
sw = set(STOPWORDS)
sw.add('진짜')
sw.add('너무')
sw.add('정말')
sw.add('영화')
sw.add('ㅋㅋ')
sw.add('그리고')
```

```python
wc=WordCloud(font_path="c:/Windows/Fonts/batang.ttc", 
             background_color="white",
             max_words=3000, 
             mask=aladdin_mask,
             stopwords = sw)
wc=wc.generate(text)
wc.recolor(color_func=color_func)
```

```python
plt.figure(figsize=(20,20))
plt.imshow(wc)
plt.axis('off')
```
