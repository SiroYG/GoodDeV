## 네이버정보처리기사 검색 후 파워링크 수집

```python
"""요청"""
url='https://search.naver.com/search.naver?query=정보처리기사'
res=req.get(url)
res
```

```python
from bs4 import BeautifulSoup as bs
```

```python
soup = bs(res.text,'html.parser')
```

```python
soup.find('a',class_='lnk_tit')
```

```python
"""find_all == findAll"""

soup.findAll('a', class_='lnk_tit')[0].text
```

```python
result = soup.findAll('a', class_='lnk_tit')
for i in range(len(result)):
    print(result[i].text)
```

```python
for i in result:
    print(i.text)
```

```python
soup.select('a.lnk_tit')#select = findAll
```

```python
soup.select_one('a.lnk_tit')#select_one = find
```

#### 기사 제목 과 내용 출력하기

```python
url = 'https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=103&oid=052&aid=0001324462'

res = req.get(url)
res
```

```python
soup = bs(res.text,'html.parser')
scr = soup.find_all('script')
for s in scr :
    s.extract()
```

```python
title = soup.select_one('#articleTitle')
title2 = soup.find('h3', id='articleTitle').text
contents = soup.select_one('#articleBodyContents')
```

```python
print('제목 : ',title.text)
print('내용 : ',contents.text.strip())
contents.text.replace('\n','').replace('\t','') #replace(이걸,이걸로바꾼다구!)
```

#### 검색순위 출력하기

```python
url = 'https://www.naver.com/'

res = req.get(url)
res
```

```python
soup = bs(res.text,'html.parser')
text = soup.select('.ah_k', limit=20)
for i in range(len(text)) :
    print(i+1,'. ',text[i].text) #print('{}. {}'.format(i+1,text[i].text))
```
